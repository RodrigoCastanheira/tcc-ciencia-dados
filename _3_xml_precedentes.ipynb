```python
# OBJETIVO:
# A partir de arquivos PDF baixados pelo notebook _2_scraping_acordaos, contendo o
# inteiro teor de acórdãos precedentes de súmulas, obter o documento XML embutido no PDF
# (quando houver), extrair os dados de interesse do XML e gerar o dataset "precedentes".
```

```python
import os
import pandas as pd
import re
import subprocess
import xml.etree.ElementTree as ET
```

```python
# obtém a relação de arquivos PDF de precedentes de súmulas
path_to_pdf = 'pdf/'
pdf_names = [name for name in os.listdir(path_to_pdf) if name.lower().endswith('.pdf')]
len(pdf_names)
```

```python
%%time
# extrai o anexo 'documento.xml' dos arquivos PDF por meio do utilitário de
# linha de comando 'pdfdetach' (https://www.xpdfreader.com/download.html)

# pasta onde se localiza o utilitário 'pdfdetach'
path_to_utils = 'utils/'

# pasta onde armazenar os anexos XML
path_to_xml = 'xml/'
os.makedirs(path_to_xml, exist_ok=True)

# lista de arquivos PDF sem qualquer anexo
pdfs_sem_anexo = []
# lista de arquivos PDF com relação de anexos inesperada
pdfs_anexo_inesperado = []

# lista de arquivos XML extraídos
xml_names = []

for pdf_name in pdf_names:
    print('Processando', pdf_name)
    # verifica se contém o anexo 'documento.xml'
    output = subprocess.run([path_to_utils + 'pdfdetach', '-list', path_to_pdf + pdf_name],
                            check=True, capture_output=True, text=True)
    if '1: documento.xml' in output.stdout:
        # novo nome para o anexo
        xml_name = pdf_name[:-3] + 'xml'
        # extrai o anexo nº 1 (documento.xml) atribuindo um novo nome
        subprocess.run([path_to_utils + 'pdfdetach', '-save', '1', '-o', path_to_xml + xml_name,
                        path_to_pdf + pdf_name], check=True)
        xml_names.append(xml_name)
    elif '0 embedded files' in output.stdout:
        pdfs_sem_anexo.append(pdf_name)
    else:
        pdfs_anexo_inesperado.append(pdf_name)
        
print('Arquivos PDF:', len(pdf_names))
print('Arquivos XML:', len(xml_names))
print('Arquivos PDF sem o anexo "documento.xml":', len(pdfs_sem_anexo))
print('Arquivos PDF com relação de anexos inesperada":', len(pdfs_anexo_inesperado))
if pdfs_anexo_inesperado:
    print('Exemplo de PDF com relação de anexos inesperada:', pdfs_anexo_inesperado[0])
```

```python
# obtém o conteúdo da ementa, relatório e voto do acórdão precedente, a partir do XML

# dicionário a ser posteriormente convertido em DataFrame
acordaos = dict()
print("Arquivos XML:", len(xml_names))
for index, xml_name in enumerate(xml_names):
    print('Processando', xml_name)
    # faz o parsing do XML
    root = ET.parse(path_to_xml + xml_name).getroot()
    
    # obtém o conteúdo de interesse
    acordao = root.findtext('Decisao')
    relatorio = root.findtext('Relatorio')
    voto = root.findtext('Voto')
    voto_vencedor = root.findtext('VotoVencedor')
    # a tag 'Ementa' pode não aparecer, p. ex., quando o recurso não é conhecido
    if root.find('Ementa') is None:
        ementa = ''
    else:
        ementa = ''.join(root.find('Ementa').itertext())

    # ignora o registro se o número do acórdão no XML não corresponder ao nome do arquivo
    if re.sub(r'\D', '', acordao) != re.sub(r'\D', '', xml_name[:-3]):
        print('Número de acórdão inesperado:', acordao, xml_name)
        continue
    
    # constrói o dicionário {index : [acordao, ementa, relatorio, voto, voto_vencedor]}
    acordaos[index] = [acordao, ementa, relatorio, voto, voto_vencedor]
    
print('Acórdãos extraídos:', len(acordaos))
```

```python
# converte o dicionário em DataFrame e exporta o dataset "precedentes"
df = pd.DataFrame.from_dict(acordaos, orient='index',
                            columns=("Acordao", "Ementa", "Relatorio", "Voto", "VotoVencedor"))
df.to_csv('datasets/precedentes.csv', index=False)
print('Precedentes extraídos:', len(df))
```
